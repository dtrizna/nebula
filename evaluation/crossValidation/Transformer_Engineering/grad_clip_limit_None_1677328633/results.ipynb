```python
import sys
sys.path.append("../../../..")
from nebula.evaluation import read_cv_metrics_file, read_cv_metrics_folder
from nebula.misc.plots import plot_cv_metrics_dict

import os

training_files = [x for x in os.listdir() if x.endswith("training.json")]
validation_files = [x for x in os.listdir() if x.endswith("validation.json")]

def training_filter(file):
    return file.endswith("training.json")
def validation_filter(file):
    return file.endswith("validation.json")
def key_extractor(file):
    return "_".join(file.split("_")[0:3])

metrics_train = read_cv_metrics_folder('.', key_extractor, training_filter)
metrics_val = read_cv_metrics_folder('.', key_extractor, validation_filter)
```
