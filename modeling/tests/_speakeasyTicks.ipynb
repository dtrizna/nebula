```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
def plotSeabornBarplotTicks(ticks):
    ticks = np.array(ticks)
    ticksMean = np.mean(ticks, axis=0)
    ticksStd = np.std(ticks, axis=0)
    ticksMean = pd.DataFrame(ticksMean, columns=["mean"])
    ticksMean["std"] = ticksStd
    ticksMean["function"] = ["getRecordsFromFile", "normalizeTableIP", "normalizeTablePath", "joinSpeakEasyRecordsToJSON", "write to file"]
    fig,ax = plt.subplots(figsize=(14,5))
    sns.barplot(x="function", y="mean", data=ticksMean, yerr=ticksMean["std"], ax=ax)

ticks = pickle.load(open("_speakeasyTicks.pkl", "rb"))
plotSeabornBarplotTicks(ticks)
```
