```python
import os
import sys
sys.path.append(".")
sys.path.append("..\\..")
from nebula.misc import getRealPath
import time
import re
from nebula.constants import SPEAKEASY_SUBFILTER_MINIMALISTIC

def flattenList(l):
    return [item for sublist in l for item in sublist]
STOPWORDS = flattenList([SPEAKEASY_SUBFILTER_MINIMALISTIC[x] for x in SPEAKEASY_SUBFILTER_MINIMALISTIC])

PATH = getRealPath(type="notebook")
ROOT = PATH + r"\..\.."
TRAINSET = ROOT + r"\data\data_filtered\speakeasy_trainset"

print(TRAINSET)
subFolders = [x for x in os.listdir(TRAINSET) if x.startswith("report_")]

def cleanJsonEvent(jsonEvent, patternCleanup = ['"', ":", ",", "[", "]", "{", "}", "\\", "/"]):
    jsonEvent = jsonEvent.lower()
    for x in patternCleanup:
        jsonEvent = jsonEvent.replace(x, " ")
    return jsonEvent
    
LIMIT = 100
jsonEvents = []
jsonEventRaw = []
for subFolder in subFolders[0:1]:
    fullPath = f"{TRAINSET}\\{subFolder}"
    files = [f"{fullPath}\\{x}" for x in os.listdir(fullPath) if x.endswith(".json")]
    l = len(files)
    for i,file in enumerate(files):
        print(f"{subFolder:>20}: {i+1}/{l} {' '*30}", end="\r")
        with open(file, "r") as f:
            jsonEvent = f.read()
        jsonEventRaw.append(jsonEvent)
        
        jsonEvent = cleanJsonEvent(jsonEvent)
        jsonEvents.append(jsonEvent)
        
        if LIMIT and i == LIMIT:
            break
```

```python
# import wordpunct tokenizer
from nltk.tokenize import WordPunctTokenizer, WhitespaceTokenizer
#tokenizer = WordPunctTokenizer()
tokenizer = WhitespaceTokenizer()

jsonEventsTokens = []
for jsonEvent in jsonEvents:
    tokenizedJsonEvent = tokenizer.tokenize(jsonEvent)
    # clean from STOPWORDS
    tokenizedJsonEvent = [x for x in tokenizedJsonEvent if x not in STOPWORDS]
    jsonEventsTokens.append(tokenizedJsonEvent)

# counter on jsonEventsTokens
from collections import Counter
counter = Counter()
for jsonEventTokens in jsonEventsTokens:
    counter.update(jsonEventTokens)
```
