```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

def plot_list_element_lengths(list_of_strings, xlim=None):
    lengths = [len(x) for x in list_of_strings]
    fix, ax = plt.subplots(1, 1, figsize=(10, 8))
    ax.hist(lengths, bins=100)
    plt.title("Histogram of list element lengths")
    plt.xlim(xlim)
    plt.show()
    return lengths
```

```python
# event char limit enforced to 20k
with open("auditd_parsed_noLengthLimit.out") as f:
    data = f.readlines()

b = plot_list_element_lengths(data)

print(f"Mean: {np.mean(b):.2f}")
print(f"Max: {np.max(b)}") # some event done A LOT of stuff during 5 minutes!!
# print 25, 50, 75 percentiles
print(np.percentile(b, [25, 50, 75]))

percentiles = []
for i in range(50,99):
    percentiles.append(np.percentile(b, i))    
# lineplot percentiles
fix, ax = plt.subplots(1, 1, figsize=(10, 8))
_ = sns.lineplot(x=range(50,99), y=percentiles, ax=ax)
```
