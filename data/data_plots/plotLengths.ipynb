```python
import sys
sys.path.append('../..')
from nebula.plots import plotListElementLengths

import json
with open("../data_filtered/speakeasy_tokenized.json") as f:
    data = json.load(f)
```

```python
import matplotlib.pyplot as plt

def plotListElementLengths(list_of_strings, xlim=None, outfile=None, bins=100, ax=None):
    lengths = [len(x) for x in list_of_strings]
    if ax is None:
        fig, ax = plt.subplots(1, 1, figsize=(10, 8))
    # plot ax.hist with bar borders
    ax.hist(lengths, bins=bins, histtype="bar", ec="black")
    ax.set_title("Histogram of list element lengths")
    if xlim:
        ax.set_xlim(xlim)
    ax.set_yscale("log")
    ax.set_ylabel("Count (log)")
    ax.grid()#which="both")
    if outfile:
        plt.tight_layout()
        plt.savefig(outfile)
    return lengths, ax

l, ax = plotListElementLengths(data, xlim=(0, 7e3), bins=700, outfile="plotLengths.png")
```
